```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def function1():
  x = np.array([i for i in range(-10,10)])
  y = x**3/12 + x * (x-15) - 72

  plt.plot(x, y)

function1()
```

```
def function2():
  x = np.array([i for i in np.arange(-10,10,0.1)])
  y = x**3/12 + x * (x-15) - 72

  plt.plot(x, y, color='r')

function2()
```

```
def f4(x):
  y = x*x*np.sin(x/100) + 300*x
  return y

def function3(f, a, b):
  x = np.array([i for i in np.arange(a,b,0.1)])
  y = [f(i) for i in x]
  
  zero = []
  
  plt.plot(x, y)

  for i in range(len(y)-1):
    if y[i] >= 0 and y[i+1] < 0:
      zero.append(i)
    if y[i] <= 0 and y[i+1] > 0:
      zero.append(i)

  plt.plot([x[i] for i in zero],[y[i] for i in zero],'gs')

function3(f4,-1000,1000)
```

```
def get_deriviate(f, a, b):
  #Вычисляем точки по x
  x = np.array([i for i in range(a,b)])
  len_x = x.shape[0]

  #Рассчитываем производную
  dy = np.array([f(x[i+1]) - f(x[i]) for i in range(len_x-1)])

  #Возвращаем производную
  return dy

def function4():
  x = np.arange(-1000,1000)
  dy = get_deriviate(f4,-1000,1001)

  dy_min = []
  dy_max = []

  min_y = min(dy)
  max_y = max(dy)

  for i in range(len(dy)):
    if dy[i] == min_y:
      dy_min.append(i)
    if dy[i] == max_y:
      dy_max.append(i)

  plt.plot(x,dy)

  plt.plot([x[i] for i in dy_min],[dy[i] for i in dy_min],'go')

  plt.plot([x[i] for i in dy_max],[dy[i] for i in dy_max],'ro')

function4()
```
